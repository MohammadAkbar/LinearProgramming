**Q2 Linear Programming Duality**

n : number of guests<br>
m : items in gift registry<br>
We want to maximize the the number of gifts recieved ie the flow. If a guest $g_x$ has an item $i_y$ on their preference list we add an edge from that guest to that item $g_x \rightarrow i_y$. Each guest node has a capacity of 1 so they will buy at most one item. All preferences are ranked equaly so all edges have a capacity of 1. The max flow of the graph will correspond to the max number of gifts. Edges from $g_x \rightarrow i_y$ with flow of 1 correspond to $g_x$ selected gift.<br>
